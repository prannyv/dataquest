```python
import pandas as pd
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

# from sklearn.metrics import accuracy_score
# from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('hotel_train.csv')
```

```python
x = pd.get_dummies(df.drop(['BookingID', 'BookingStatus','MealPlan'], axis=1))
```

```python
y = df['BookingStatus'].apply(lambda i: 1 if i=='Canceled' else 0)
```

```python
model = Sequential()
model.add(Dense(units=32,activation='relu',input_dim=len(x.columns)))
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=64,activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='Adam', metrics='accuracy')
```

```python
model.fit(x,y,epochs=300,batch_size=32)
```

```python
newDf = pd.read_csv('hotel_test.csv')
test = pd.get_dummies(newDf.drop(['BookingID', 'BookingStatus', 'MealPlan'], axis=1))
#bug w the meal plan
prediction = model.predict(test)
prediction = [0 if val < 0.5 else 1 for val in prediction]
```
